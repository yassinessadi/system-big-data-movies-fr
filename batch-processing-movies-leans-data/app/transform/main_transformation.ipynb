````python
from sqlalchemy import String
from sqlalchemy.orm import DeclarativeBase
from sqlalchemy.orm import Mapped
from sqlalchemy.orm import mapped_column
from sqlalchemy import create_engine


class Base(DeclarativeBase):
    pass

class Genres(Base):
    __tablename__ = "Genres"
    id: Mapped[int] = mapped_column(primary_key=True)
    name: Mapped[str] = mapped_column(String(30))
    def __repr__(self) -> str:
        return f"Genres(id={self.id!r}, name={self.name!r}"



def read_files(prefix:str):
    """
    Reads files based on the specified prefix.
    Example: cast | crew | movie_cast | movie_crew | movies
    """
    now = datetime.now()
    file_name = now.strftime("%Y-%m-%d")
    file_name = f'../../test/{prefix}_{file_name}.txt'
    response = pd.read_json(file_name, lines=True)
    return response


genres = read_files(prefix='genres')


def insertData(database_url:str):
    engine = create_engine(database_url, echo=True)
    # Create the table in the database
    Base.metadata.create_all(bind=engine)
    engine.connect()
    genres.to_sql('Genres',con=engine,index=False, if_exists='replace')
    engine.dispose()

insertData("mssql+pyodbc://jane_essadi:123123@192.168.208.1/TheMoviesDB?driver=ODBC+Driver+17+for+SQL+Server")
````
